# Data analysis of *Hacker News*

Two different types of posts `Ask HN` and `Show HN` from ([Hacker News](https://news.ycombinator.com/)) will be compared in this project. ([Hacker News](https://news.ycombinator.com/)) is an extremely popular webpage in technology and startup circles.

`Ask HN` is a page for users to submit posts asking the community aspecific question, such as "What would a Facebook that isn't evil look like?" Likewise, a project, product or something interesting can be submitted on `Show HN`.

We will try to find an answer to these two questions:

- Do `Ask HN` or `Show HN` receive more comments on average?
- Do posts created at a certain time receive more comments on average?

We will be working with the data set reduced from almost 300,000 rows to approximately 20,000 rows by removing all submissions that did not receive any comments, and then randomly sampling from the remaining submissions.

## Introduction

First of all, we will read in the data and remove the headers.

```python
# Read in the data

from csv import reader
open_file = open('hacker_news.csv')
read_file = reader(open_file)
hn = list(read_file)
```

```python
print(hn[:5])
```

## Removing headers from a list of lists

```python
# Remove the headers

headers = hn[:1]
hn = hn[1:]
print(headers)
print(hn[:5])
```

We can see above the data set contains the title of the posts, the number of comments, and the date the post was submitted. The number of comments will be explored below.

## Extracting `Ask HN` and `Show HN` posts

First, the posts begin with either `Ask HN` or `Show HN` will be identified and we will then separate the data for those  two types of posts into different lists. Separating the data makes it easier to analyze in the following steps.

```python
# Identify posts in terms of `Ask HN` or `Show HN` and separate the data into different lists

ask_posts = []
show_posts = []
other_posts = []

for row in hn:
    title = row[1]
    title = title.lower()
    if title.startswith('ask hn'):
        ask_posts.append(row)
    elif title.startswith('show hn'):
        show_posts.append(row)
    else:
        other_posts.append(row)
```

```python
print(len(ask_posts))

print(len(show_posts))

print(len(other_posts))
```

```python
print(ask_posts[:5])

print(show_posts[:5])
```

## Average number of comments for `Ask HN` and `Show HN` posts

Now we will calculate the avergae number of comments based on different lists.

```python
# Calculate the avergare number of comments `Ask HN` posts receive

total_ask_comments = 0

for row in ask_posts:
    total_ask_comments += int(row[4])

avg_ask_comments = total_ask_comments / len(ask_posts)
print(avg_ask_comments)
print(total_ask_comments)
  

    
    
```

```python
# Calculate the avergare number of comments `Show HN` posts receive

total_show_comments = 0

for row in show_posts:
    total_show_comments += int(row[4])
    
avg_show_comments = total_show_comments / len(show_posts)
print(avg_show_comments)
print(total_show_comments)
```

On average, show posts receive approximately **10**, whereas ask posts receive about **14** comments. Ask posts will be our main focus for analysis, since this posts are likely to receive more comments. 

## The amount of Ask posts and comments by hour

First, the amount of ask posts during each hour of day will be identified, along with the number of comments. Then, the average amount of comments ask posts will be calculated at each hour of the day.

```python
# Calculate the amount of ask posts created during each hour of day and the number of comments received

import datetime as dt

result_list = []

for row in ask_posts:
    created_at = row[6]
    num_comments = int(row[4])
    result_list.append([created_at, num_comments])
    print(result_list[:6])
    
    
```

```python
counts_by_hour = {}
comments_by_hour = {}

for row in result_list:
    date = dt.datetime.strptime(row[0], "%m/%d/%Y %H:%M")
    hour = date.strftime("%H")
    if hour not in counts_by_hour:
        counts_by_hour[hour] = 1
        comments_by_hour[hour] = row[1]
    else:
        counts_by_hour[hour] += 1
        comments_by_hour[hour] += row[1]
```

```python
print(counts_by_hour)
print(comments_by_hour)
```

## The average number of comments for `Ask HN` posts by hour

```python
# Calculate the average amount of comments `Ask HN` posts created at each hour of the day receive

hours = []

for hour in comments_by_hour:
    hours.append([hour, comments_by_hour[hour] / counts_by_hour[hour]])
  
```

```python
print(hours)
```

## Sorting and printing values

```python
swap_avg_by_hour = []

for row in hours:
    swap_avg_by_hour.append([row[1],row[0]])
   
    

```

```python
from operator import itemgetter

sorted_swap = sorted(swap_avg_by_hour, key=itemgetter(0), reverse=True)
print(sorted_swap)
```

```python
# Sort the values and print the 5 hours with the highest average comments

print("Top 5 Hours for Ask Posts Comments:")
hour_format ="%H"
for row in sorted_swap[:5]:
    hour = row[1]
    avg = row[0]
    
    convert_to_cet = dt.datetime.strptime(hour, hour_format)
    cet = convert_to_cet + dt.timedelta(hours=6)
    cet = cet.strftime("%H:%M")
    
    est = convert_to_cet.strftime("%H:%M")
    
    
    
    print("\t")
    
    print("{0}: {1:.2f} average comments per post (CET)".format(cet, avg))
    
    print("\t")
    
    print("{0}: {1:.2f} average comments per post (EST)".format(est, avg))
    
    
```

Eastern time zone is also converted to Central Eauropean time above to make it easier for each region to understand the average comments per posts.

The hour that receives the most comments per post on average is **21:00 (CET)**, with an average of **38.59** comments per post. There's about a **60%** increase in the number of comments between the hours with the highest and second highest average number (**23.81**) of comments. The lowest average number of comments is **16**, and **2.4** times less than the highest average number of comments.
